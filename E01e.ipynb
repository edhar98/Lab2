```python
R_l = [] # the list of resistances (Ohm)

# Instantiate the Resistances for the setup
R3 = 200
R2 = 5000
R1 = 10000

R_l.append(R1)
R_l.append(R2)
R_l.append(R3)

U_ps = 5 # Voltage of the power supply for DC
```

```python
def total_resistance(R: list, circuit_type: str):
    if circuit_type == 'p': # parallel
        tmp_sum = 0
        for r in R:
            tmp_sum += 1/r
        return 1/tmp_sum
    if circuit_type == 's': # serial
        return sum(R)
    return None
```

```python
def I_dc(R: float):
    return U_ps / R # Amp
```

```python
R_s = total_resistance(R_l, 's')
```

```python
I=I_dc(R_s)
```
